```julia
variance = [2^n for n in 0:8]
```

```julia
bias_squared = reverse(variance)
```

```julia
total_error = variance + bias_squared
```

```julia
plot!(bias_squared, label = "bias^2", color = "red", lienstyle = :)
```

```julia
plot(variance, label="variance", color="green", linestyle= :solid)
```

```julia
plot!(bias_squared, label="bias^2", color="red", linestyle= :dashdot)
```

```julia
plot!(total_error, label="total error", color="blue", linestyle = :dot)
```

```julia
xlabel!("model complexity")
```

```julia
title!("the Bias-Variance Tradeoff")
```

```julia
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
```

```julia
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
```

```julia
labels = 'a':'i'
```

```julia
scatter(friends, minutes)
```

```julia
for (label, friend_count, minute_count) in zip(labels, friends, minutes)
    annotate!(friend_count + 0.25, minute_count, text(label, :left))
end
```

```julia
current()
```

```julia
title!("Daily Minutes vs. Number of Friends")
```
